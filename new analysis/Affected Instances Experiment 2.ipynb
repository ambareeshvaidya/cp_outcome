```python
## Features data set
# import set1,2 for ml
ml = pd.read_csv('experiment_2_affected_analysis.csv')
ml = ml.drop(['Unnamed: 0', 'RANDOM SEED', 'TIME ROOT END', 'ROUND OF CUTS', 'OBJECTIVE VALUE ROOT NODE'], axis = 1)
ml
```

```python
df_train_1 = pd.read_csv('experiment_2_train.csv')
df_train_1 = df_train_1.filter(['NAME', 'TIME W/ CUTS', 'TIME W/O CUTS'], axis = 1)
df_train_1
```

```python
affected = []
for index, row in df_train_1.iterrows():
    
    if row['TIME W/ CUTS'] < 0.8*row['TIME W/O CUTS'] or row['TIME W/O CUTS'] < 0.8*row['TIME W/ CUTS']:
        affected.append(1)
        
    else:
        affected.append(0)
        
        
df_train_1['Affected'] = affected
df_train_1['Affected'].value_counts()[1]
df_train_1 = df_train_1.drop(['TIME W/ CUTS', 'TIME W/O CUTS'], axis = 1)
df_train_1

set_1_train = df_train_1.merge(ml, on = ['NAME'], how = 'left')
set_1_train
```

```python
reps = [10 if val==1 else 1 for val in df_train_1.Affected]
df_train_2 = df_train_1.loc[np.repeat(df_train_1.index.values, reps)]
df_train_2

set_2_train = df_train_2.merge(ml, on = ['NAME'], how = 'left')
set_2_train
```

```python
df_train_3 = df_train_1.loc[df_train_1['Affected'] == 1]
df_train_3

set_3_train = df_train_3.merge(ml, on = ['NAME'], how = 'left')
set_3_train
```

```python
df_train_3_unaff = df_train_1.loc[df_train_1['Affected'] == 0]
df_train_3_unaff = df_train_3_unaff.filter(['NAME', 'Affected'], axis = 1)
df_train_3_unaff

set_3_train_unaff = df_train_3_unaff.merge(ml, on = ['NAME'], how = 'left')
set_3_train_unaff
```

```python
df_test = pd.read_csv('experiment_2_test.csv')
df_test = df_test.filter(['NAME', 'TIME W/ CUTS', 'TIME W/O CUTS'], axis = 1)
df_test
```

```python
affected = []
for index, row in df_test.iterrows():
    
    if row['TIME W/ CUTS'] < 0.8*row['TIME W/O CUTS'] or row['TIME W/O CUTS'] < 0.8*row['TIME W/ CUTS']:
        affected.append(1)
        
    else:
        affected.append(0)
        
        
df_test['Affected'] = affected
df_test['Affected'].value_counts()[1]
df_test
```

```python
df_test = df_test.drop(['TIME W/ CUTS', 'TIME W/O CUTS'], axis = 1)
test = df_test.merge(ml, on = ['NAME'], how = 'left')
test
```

SETUP 1

```python
X_train = set_1_train.drop(['NAME', 'Affected', 'TIME W/ CUTS', 'TIME W/O CUTS', 'NODES W/ CUTS', 'NODES W/O CUTS', 
                            'Time Label', 'constr', 'linbinpack_constr', 'PRESOLVING TIME W/O CUTS', 
                            'linmixbin_constr', 'cont_vars', 'presol_int_vars', 'PRESOLVING TIME W/ CUTS'], axis = 1)
X_test = test.drop(['NAME', 'Affected', 'TIME W/ CUTS', 'TIME W/O CUTS', 'NODES W/ CUTS', 'NODES W/O CUTS', 
                    'Time Label', 'constr', 'linbinpack_constr', 'PRESOLVING TIME W/O CUTS', 'linmixbin_constr', 
                    'cont_vars', 'presol_int_vars', 'PRESOLVING TIME W/ CUTS'], axis = 1)

y_train = set_1_train.filter(['Time Label'], axis = 1)
y_test = test.filter(['Time Label'], axis = 1)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
X_train
```

```python
# Scaling data
scaler = StandardScaler()
# Fit only to the training data
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print('RANDOM FOREST')
clf = RandomForestClassifier(random_state = 0, max_depth = 10, n_estimators = 200)
clf.fit(X_train, y_train)
y_train_pred_rf = clf.predict(X_train)
y_test_pred_rf = clf.predict(X_test)

score_test = accuracy_score(y_test, y_test_pred_rf)
score_train = accuracy_score(y_train, y_train_pred_rf)

print("Accuracy score test: {}\n".format(score_test))
print("Accuracy score train: {}\n".format(score_train))
```

```python
set_1_train['Pred'] = y_train_pred_rf
set_1_train_affected = set_1_train.loc[set_1_train['Affected'] == 1]
# set_1_train_affected
test['Pred'] = y_test_pred_rf
test_affected = test.loc[test['Affected'] == 1]
# test_affected

train_time_label = set_1_train_affected['Time Label'].values.tolist()
train_pred_label = set_1_train_affected['Pred'].values.tolist()

test_time_label = test_affected['Time Label'].values.tolist()
test_pred_label = test_affected['Pred'].values.tolist()

score_train_affected = accuracy_score(train_time_label, train_pred_label)
score_test_affected = accuracy_score(test_time_label, test_pred_label)

print("Accuracy score test affected: {}\n".format(score_test_affected))
print("Accuracy score train affected: {}\n".format(score_train_affected))
```
