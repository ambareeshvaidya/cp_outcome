```python
import pandas as pd
import numpy as np
```

```python
# cuts-off dataframes

df_0_2 = pd.read_csv('result_0_2.csv', nrows = 426)
df_0_4 = pd.read_csv('result_0_4.csv', nrows = 426)
df_0_8 = pd.read_csv('result_0_8.csv', nrows = 426)
df_0_16 = pd.read_csv('result_0_16.csv', nrows = 426)
df_0_32 = pd.read_csv('result_0_32.csv', nrows = 426)
df_0_32
```

```python
df_nocuts = pd.concat([df_0_2,df_0_4,df_0_8,df_0_16,df_0_32])
df_nocuts = df_nocuts.sort_values(by='NAME', ascending=True)
```

```python
df_nocuts.rename(columns={'INITIAL LP TIME':'PRESOLVING TIME W/O CUTS'}, inplace = True)
df_nocuts = df_nocuts.drop(['PRESOLVING TIME'], axis = 1)
df_nocuts.to_csv('df_nocuts.csv')
```

```python
# cuts-on data

df_1_2 = pd.read_csv('result_1_2.csv', nrows = 426)
df_1_4 = pd.read_csv('result_1_4.csv', nrows = 426)
df_1_8 = pd.read_csv('result_1_8.csv', nrows = 426)
df_1_16 = pd.read_csv('result_1_16.csv', nrows = 426)
df_1_32 = pd.read_csv('result_1_32.csv', nrows = 426)
df_1_32
```

```python
df_cuts = pd.concat([df_1_2,df_1_4,df_1_8,df_1_16,df_1_32])
df_cuts = df_cuts.sort_values(by='NAME', ascending=True)
```

```python
df_cuts.rename(columns={'INITIAL LP TIME':'PRESOLVING TIME W/ CUTS'}, inplace = True)
df_cuts = df_cuts.drop(['PRESOLVING TIME'], axis = 1)
df_cuts.to_csv('df_cuts.csv')
```
