```python
# import result sets

data_no_cuts = pd.read_csv("results_0.csv")
data_cuts = pd.read_csv("results_1.csv")

max_repeat = 5
vc = data_no_cuts['NAME'].value_counts()
c = data_cuts['NAME'].value_counts()
data_no_cuts = data_no_cuts[data_no_cuts['NAME'].isin(vc[vc == max_repeat].index)]
data_cuts = data_cuts[data_cuts['NAME'].isin(c[c == max_repeat].index)]

# data_no_cuts
```

```python
data_no_cuts_new = data_no_cuts.loc[data_no_cuts.NAME.isin(data_cuts.NAME)]
data_cuts_new = data_cuts.loc[data_cuts.NAME.isin(data_no_cuts.NAME)]

print(data_cuts_new.shape)
print(data_no_cuts_new.shape)
```

```python
nc_timeout = data_no_cuts_new[(data_no_cuts_new["SOLUTION TIME"] >= 18000)]
c_timeout = data_cuts_new[(data_cuts_new["SOLUTION TIME"] >= 18000)]


max_repeat = 5
vc = nc_timeout['NAME'].value_counts()
c = c_timeout['NAME'].value_counts()
nc_timeout = nc_timeout[nc_timeout['NAME'].isin(vc[vc == max_repeat].index)]
c_timeout = c_timeout[c_timeout['NAME'].isin(c[c == max_repeat].index)]

nc_timeout = nc_timeout.filter(['NAME', 'SOLUTION TIME'])
c_timeout = c_timeout.filter(['NAME', 'SOLUTION TIME'])

# print(nc_timeout.head(50))
# print(nc_timeout.shape)
# print(c_timeout.shape)

nc_timeout = nc_timeout.groupby('NAME').mean()
c_timeout = c_timeout.groupby('NAME').mean()

print(nc_timeout.shape)
print(c_timeout.shape)

timeout = nc_timeout.merge(c_timeout, on = "NAME", how = "left")
timeout.info()
# timeout.to_csv('analysis.csv')
timeout = timeout.dropna()
timeout
```

```python
name_list = timeout.index.values.tolist()
name_list

no_cuts = data_no_cuts_new[~data_no_cuts_new.NAME.isin(name_list)]
cuts = data_cuts_new[~data_cuts_new.NAME.isin(name_list)]
print(len(cuts))
print(len(no_cuts))
```

```python
final = pd.concat([cuts, no_cuts])
final = final.sort_values(by = ['NAME', 'CUT SETTING'], ascending = True)
final
# final.to_csv('final.csv')
```

```python
static = pd.read_csv("features_original.csv")
static = static.filter(['instance_name', 'vars', 'bin_vars', 'int_vars', 'cont_vars', 
                       'constr', 'linsetpart_constr', 'linsetpack_constr', 'linsetcov_constr', 'lincard_constr',
                       'lineqknap_constr', 'linbinpack_constr', 'linknaps_constr', 'linintknap_constr',
                        'linmixbin_constr', 'nonzeroes'], axis = 1)
static['instance_name'] = static.instance_name.str.split('/').str[-1]

static_new = static.loc[static.instance_name.isin(final.NAME)]

static_new['nonzeroes'] = static_new['nonzeroes'] / (static_new['vars'] * static_new['constr'] )

static_new['bin_vars'] = static_new['bin_vars'] / static_new['vars']
static_new['int_vars'] = static_new['int_vars'] / static_new['vars']
static_new['cont_vars'] = static_new['cont_vars'] / static_new['vars']

static_new['linbinpack_constr'] = static_new['linbinpack_constr'] / static_new['constr']
static_new['lincard_constr'] = static_new['lincard_constr'] / static_new['constr']
static_new['lineqknap_constr'] = static_new['lineqknap_constr'] / static_new['constr']
static_new['linintknap_constr'] = static_new['linintknap_constr'] / static_new['constr']
static_new['linknaps_constr'] = static_new['linknaps_constr'] / static_new['constr']
static_new['linmixbin_constr'] = static_new['linmixbin_constr'] / static_new['constr']
static_new['linsetcov_constr'] = static_new['linsetcov_constr'] / static_new['constr']
static_new['linsetpack_constr'] = static_new['linsetpack_constr'] / static_new['constr']
static_new['linsetpart_constr'] = static_new['linsetpart_constr'] / static_new['constr']

static_new['vars'] = np.log(static_new['vars'])
static_new['constr'] = np.log(static_new['constr'])

static_new.rename(columns={'instance_name': 'NAME'}, inplace=True)
static_new
```

```python
dynamic = pd.read_csv("features_after_trivial_presolving.csv")
dynamic = dynamic.filter(['instance_name', 'vars', 'int_vars', 'constr'], axis = 1)

dynamic['instance_name'] = dynamic.instance_name.str.split('/').str[-1]

dynamic_new = dynamic.loc[dynamic.instance_name.isin(final.NAME)]

dynamic_new['int_vars'] = dynamic_new['int_vars'] / dynamic_new['vars']

dynamic_new['vars'] = np.log(dynamic_new['vars'])
dynamic_new['constr'] = np.log(dynamic_new['constr'])

dynamic_new.rename(columns={'instance_name': 'NAME', 'vars': 'presol_vars', 'int_vars': 'presol_int_vars', 
                            'constr': 'presol_constr'}, inplace=True)
dynamic_new
```

```python
final['DIG NUM'] = abs(final['OBJECTIVE VALUE ROOT NODE'] - final['INITIAL LP']) 
final['PDG NUM'] = abs(final['BEST PRIMAL BOUND'] - final['OBJECTIVE VALUE ROOT NODE']) 
final['PIG NUM'] = abs(final['BEST PRIMAL BOUND'] - final['INITIAL LP']) 

final['DUAL INITIAL GAP'] = final['DIG NUM'] / abs(final[['OBJECTIVE VALUE ROOT NODE', 'INITIAL LP', 'DIG NUM']]).max(axis = 1)
final['PRIMAL DUAL GAP'] = final['PDG NUM'] / abs(final[['OBJECTIVE VALUE ROOT NODE', 'BEST PRIMAL BOUND', 'PDG NUM']]).max(axis = 1)
final['PRIMAL INITIAL GAP'] = final['PIG NUM'] / abs(final[['INITIAL LP', 'BEST PRIMAL BOUND', 'PIG NUM']]).max(axis = 1)

final['GAP CLOSED'] = 1 - (final['PRIMAL DUAL GAP'] / final['DUAL INITIAL GAP'])


final = final.drop(['GAP', 'No. OF CUTS GENERATED', 'STATUS', 'END', 'DIG NUM', 'PDG NUM', 'PIG NUM'], axis =1)
final
```

```python
#dataframe with all features
final_1 = final.merge(static_new, on = "NAME", how = "left")
final_2 = final_1.merge(dynamic_new, on = "NAME", how = "left")
pd.set_option('display.max_columns', None)
final_2
# final_2.to_csv('final_2.csv')
```
